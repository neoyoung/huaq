### 广播(对不同shape数组进行运算)

```python
a = np.array([1,2,3,4])
b = np.array([5,6,7,8])
c = a*b
c
```

```python
a = np.array([1,2,3,4])
#a = np.array([[1,2,3,4],[1,2,3,4]])

b = np.array([[5,6,7,8],[5,6,7,8]])
c = a*b
c
```

```python
a = np.ones((1,2,3))
b = np.ones((3,2,3))
c = a*b
c
```

### 排序筛选函数

## 文件读写

Numpy可以读写文本数据或者二进制数据

Numpy为ndarray引入了一个简单的文件格式：.npy文件，用于存放重建ndarray数据所需的信息。

np.load(file)

np.save(file,array)

np.savez(file,array1,array2,...)

np.savetxt(file,array,fmt,delimiter)

np.loadtxt(file,dtype,delimiter)

```python
np.save('/home/mw/3/year.npy',year)
```

```python
year_tmp = np.load('/home/mw/3/year.npy')
year_tmp
```

```python
np.savez('/home/mw/3/year&t.npz',year=year,temperature=temperature)
```

```python
tmp = np.load('/home/mw/3/year&t.npz')
tmp.files
tmp['year']
```

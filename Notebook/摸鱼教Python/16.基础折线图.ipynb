```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(1979,2021,1)
y1 = np.sin(2*np.pi*x)*1e12
y2 = np.sin(3*np.pi*x)*1e12
```

```python
print(x.shape,y1.shape,y2.shape)
```

```python
fig1 = plt.figure(figsize=(15,8))
ax = fig1.add_axes([0.1,0.1,0.5,0.3])
ax.plot(x,y2,ls='-',lw=3,c='green',label='PRE',marker='o',ms=10,markeredgecolor='k',
markeredgewidth=3,markerfacecolor='w',zorder=2)
ax.plot(x,y1,ls='-',lw=3,c='purple',label='SAT',marker='o',ms=10,markeredgecolor='k',
markeredgewidth=3,markerfacecolor='w',zorder=1)
ax.set_ylim(4,-4)
ax.set_yticks([4,2,0,-2,-4])
ax.set_yticklabels(['4d','2d','0d','-2d','-4d'],fontsize=16)
ax.set_title('(a)',fontsize=18,loc='left')
ax.set_title('SAT & PRE',fontsize=18,loc='center')
fig1.text(0.33,0.5,'Plot',fontsize=18)
ax.legend(loc=3,ncol=2)
```

```python
fig1 = plt.figure(figsize=(15,8))
ax = fig1.add_axes([0.1,0.1,0.5,0.3])
ax.plot(x,y1,ls='-',c='r',alpha=0.6,lw=2,marker='o',
markeredgecolor='k',markeredgewidth=2,markerfacecolor='b',ms=10,zorder=0,label='index1')
ax.plot(x,y2,ls='--',c='k',alpha=0.6,lw=2,marker='o',
markeredgecolor='k',markeredgewidth=2,markerfacecolor='g',ms=10,zorder=1,label='index2')
ax.legend()
ax.set_title('(a) Index1 & Index2',fontsize=14,loc='left')
ax.set_ylim(-3,3)
ax.axhline(0)
ax.set_xticks([1985,1995,2005,2015])
ax.set_xticklabels(['1985','1995','2005','2015'])
plt.show()
```

```python
fig1 = plt.figure(figsize=(15,8))
ax = fig1.add_axes([0.1,0.1,0.5,0.3])
ax.plot(x,y1,ls='-',c='r',alpha=0.6,lw=2,zorder=0,label='index1')
ax.plot(x,y2,ls='-',c='k',alpha=0.6,lw=2,zorder=1,label='index2')
ax.legend()
ax.set_title('(a) Index1 & Index2',fontsize=14,loc='left')
ax.set_ylim(-3,3)
ax.axhline(0,lw=10,ls='--',c='k')
ax.axvline(1995)

ax.set_xticks([1985,1995,2005,2015])
ax.set_xticklabels(['1985','1995','2005','2015'])
ax.fill_between(x, y1, y2, where=(y2 > y1), facecolor='b', alpha=0.5,interpolate=True)
plt.show()
```

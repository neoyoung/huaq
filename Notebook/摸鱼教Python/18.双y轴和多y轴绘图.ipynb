# 双y轴和多y轴绘图


## 1.双y轴绘图

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(1979,2021,1)
y1 = np.sin(2*np.pi*x)*1e12
y2 = np.sin(3*np.pi*x)*1e12*10
```

```python
fig1 = plt.figure(figsize=(15,8))
ax = fig1.add_axes([0.1,0.1,0.5,0.3])
ax.plot(x,y1,ls='-',lw=3,c='green',label='PRE',marker='o',ms=10,markeredgecolor='k',
markeredgewidth=3,markerfacecolor='w',zorder=2)
ax.plot(x,y2,ls='-',lw=3,c='purple',label='SAT',marker='*',ms=10,markeredgecolor='k',
markeredgewidth=3,markerfacecolor='w',zorder=1)
ax.legend(loc=3,ncol=2)
plt.show()
```

```python
fig1 = plt.figure(figsize=(15,8))
ax = fig1.add_axes([0.1,0.1,0.5,0.3])
ax.plot(x,y1,ls='-',lw=3,c='green',label='PRE',marker='o',ms=10,markeredgecolor='k',
markeredgewidth=3,markerfacecolor='w',zorder=2)
ax2 = ax.twinx()
# ax2.plot(x,y2,ls='-',lw=3,c='purple',label='SAT',marker='*',ms=10,markeredgecolor='k',
# markeredgewidth=3,markerfacecolor='w',zorder=1)
ax2.bar(x,y2,label='SAT')
lines, labels = ax.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax2.legend(lines+lines2,labels+labels2,loc=3,ncol=2)
plt.show()
```

## 2.多y轴绘图

```python
x = np.arange(1979,2021,1)
y1 = np.sin(2*np.pi*x)*1e12
y2 = np.sin(3*np.pi*x)*1e12*10
y3 = np.sin(4*np.pi*x)*1e12*100
y4 = np.sin(5*np.pi*x)*1e12*1000
```

```python
fig1 = plt.figure(figsize=(15,8))
ax = fig1.add_axes([0.1,0.1,0.5,0.3])
ax.plot(x,y1,ls='-',lw=3,c='green',label='PRE',marker='o',ms=10,markeredgecolor='k',
markeredgewidth=3,markerfacecolor='w',zorder=2)
ax.plot(x,y2,ls='-',lw=3,c='purple',label='SAT',marker='*',ms=10,markeredgecolor='k',
markeredgewidth=3,markerfacecolor='w',zorder=1)
ax.plot(x,y3,ls='-',lw=3,c='red',label='WIS',marker='^',ms=10,markeredgecolor='k',
markeredgewidth=3,markerfacecolor='w',zorder=1)
ax.legend(loc=3,ncol=2)
plt.show()
```

```python
from mpl_toolkits.axisartist.parasite_axes import HostAxes, ParasiteAxes

fig1 = plt.figure(figsize=(15,8))
ax = HostAxes(fig1, [0.1,0.1,0.5,0.3]) 
ax1 = ParasiteAxes(ax, sharex=ax)
ax2 = ParasiteAxes(ax, sharex=ax)
ax3 = ParasiteAxes(ax, sharex=ax)
ax.parasites.append(ax1)
ax.parasites.append(ax2)
ax.parasites.append(ax3)
fig1.add_axes(ax)
ax.axis['right'].set_visible(False)
ax1.axis['right'].set_visible(True)
ax1.axis['right'].major_ticklabels.set_visible(True)
ax1.axis['right'].label.set_visible(True)
ax.set_ylabel('PRE')
ax1.set_ylabel('SAT')
ax2.set_ylabel('WIS')
ax3.set_ylabel('SLP')
ax2.axis['right2'] = ax2.get_grid_helper().new_fixed_axis(loc='right', axes=ax2, offset=(60,0))
ax3.axis['right3'] = ax3.get_grid_helper().new_fixed_axis(loc='right', axes=ax3, offset=(120,0))
ax1.axis['right'].major_ticks.set_color('red')
ax1.axis['right'].major_ticklabels.set_color('red')
ax1.axis['right'].line.set_color('red')
ax2.axis['right2'].major_ticks.set_color('green')
ax2.axis['right2'].major_ticklabels.set_color('green')
ax2.axis['right2'].line.set_color('green')
ax3.axis['right3'].major_ticks.set_color('blue')
ax3.axis['right3'].major_ticklabels.set_color('blue')
ax3.axis['right3'].line.set_color('blue')

ax.plot(x,y1,ls='-',lw=3,c='k',label='PRE',marker='o',ms=10,markeredgecolor='k',
markeredgewidth=3,markerfacecolor='w',zorder=2)
ax1.plot(x,y2,ls='-',lw=3,c='r',label='SAT',marker='*',ms=10,markeredgecolor='k',
markeredgewidth=3,markerfacecolor='w',zorder=1)
ax2.bar(x,y3,color='g',label='WIS')
ax3.plot(x,y4,ls='-',lw=3,c='b',label='SLP',marker='^',ms=10,markeredgecolor='k',
markeredgewidth=3,markerfacecolor='w',zorder=1)
ax3.set_ylim(-5000,8000)

ax.legend()
plt.show()
```

# 柱状图

## 基础柱状图

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
htd = np.array([10,12,16,6,19])
ltd = np.array([9,11,17,4,22])
month = ['jan','feb','mar','apr','may']
```

```python
fig = plt.figure(figsize=(15,8))
ax1 = fig.add_axes([0.1,0.1,0.6,0.4])
ax1.bar(np.arange(5),htd,width=0.3,align='edge',label='htd',color='gray',edgecolor='k',linewidth=3,hatch='.')
ax1.bar(np.arange(5)-0.3,ltd,width=0.3,align='edge',label='htd',color='white',edgecolor='k',linewidth=3,hatch='/')
ax1.set_xticks(np.arange(5),month,fontsize=16)
ax1.legend()
plt.show()
```

```python
fig = plt.figure(figsize=(15,8))
ax1 = fig.add_axes([0.1,0.1,0.6,0.4])
ax1.bar(np.arange(5),htd,width=0.2,label='htd',color='gray',edgecolor='k',linewidth=3,hatch='.')
ax1.bar(np.arange(5)-0.2,ltd,width=0.2,label='htd',color='white',edgecolor='k',linewidth=3,hatch='/')
ax1.bar(np.arange(5)+0.2,ltd+2,width=0.2,label='htd',color='white',edgecolor='k',linewidth=3,hatch='/')

ax1.set_xticks(np.arange(5),month,fontsize=16)
ax1.legend()
plt.show()
```

```python
fig = plt.figure(figsize=(15,8))
ax1 = fig.add_axes([0.1,0.1,0.6,0.4])
ax1.bar(np.arange(5),htd,width=0.3,label='htd',color='gray',edgecolor='k',linewidth=3,hatch='.')
ax1.bar(np.arange(5),ltd,width=0.3,label='htd',color='white',edgecolor='k',linewidth=3,hatch='/',bottom=htd)
ax1.set_xticks(np.arange(5),month,fontsize=16)
ax1.legend()
plt.show()
```

```python
fig = plt.figure(figsize=(15,8))
ax1 = fig.add_axes([0.1,0.1,0.6,0.4])
ax1.barh(np.arange(5),htd,label='htd',color='gray',edgecolor='k',linewidth=3,hatch='.')
ax1.set_yticks(np.arange(5),month,fontsize=16)
ax1.legend()
plt.show()
```

```python
htd_err = np.array([10,12,16,6,19])/10
ltd_err = np.array([9,11,17,4,22])/10
```

```python
fig = plt.figure(figsize=(15,8))
ax1 = fig.add_axes([0.1,0.1,0.6,0.4])
bar1 = ax1.bar(np.arange(5),htd,width=0.3,align='edge',label='htd',color='gray',edgecolor='k',linewidth=3,hatch='.',yerr=htd_err,ecolor='r')
bar2 = ax1.bar(np.arange(5)-0.3,ltd,width=0.3,align='edge',label='htd',color='white',edgecolor='k',linewidth=3,hatch='/',yerr=ltd_err,ecolor='r')
ax1.set_xticks(np.arange(5),month,fontsize=16)
ax1.legend()
ax1.bar_label(bar1,labels= np.array(['*','','','**','*']),color='r',fontsize=18)
ax1.bar_label(bar2)
plt.show()
```

# 2.确定极端事件的分布

## 2.1极端事件指数(绝对)

```python
print(t2m_array.shape)
days=0
for i in range(t2m_array.shape[0]):
    if(t2m_array[i]>28):
        days+=1
print(days)

# days = (t2m_array>28).sum()
days = np.sum(t2m_array>28)

print(days)
```

```python
t2m_yearly = t2m_array.reshape((31,31))
days = np.zeros((31))
for y in range(t2m_yearly.shape[0]):
    for d in range(t2m_yearly.shape[1]):
        if (t2m_yearly[y,d]>28):
            days[y]+=1
print(days)

days = (t2m_yearly>28).sum(axis=1)
print(days)
```

# 3.确定极端事件的时间

```python
f_t2m = xr.open_dataset('/home/mw/input/moyu1828/t1980_2010.nc')
t2m = f_t2m.air.loc[f_t2m.time.dt.month.isin(8),39.047,120] - 273.15
time = t2m.time
```

```python
date = time[t2m>28]
print(date)
```

```python
from datetime import datetime, timedelta
from pandas.core.frame import DataFrame 

date = []
for y in range(t2m_yearly.shape[0]):
    for d in range(t2m_yearly.shape[1]):
        if (t2m_yearly[y,d]>28):
            date.append(datetime(year=(1980+y),month=8,day=(d+1)))

date = DataFrame({"date":date})   
print(date)
```

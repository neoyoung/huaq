```python
x = np.arange(-3.0, 3.0, 0.2)
y = np.arange(-2.0, 2.0, 0.2)
X, Y = np.meshgrid(x, y)
Z1 = np.exp(-X**2 - Y**2)
Z2 = np.exp(-(X - 1)**2 - (Y - 1)**2)
z = (Z1 - Z2) * 2
print(x.shape,y.shape,z.shape)
```

## 2.1等值线图

https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.contour.html#matplotlib.axes.Axes.contour

```python
fig, ax = plt.subplots()
c = ax.contour(x, y, z)
ax.clabel(c)
```

## 2.3矢量图

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(0, 2 * np.pi, .2)
y = np.arange(0, 2 * np.pi, .2)
X, Y = np.meshgrid(x,y)
u = np.cos(X)
v = np.sin(Y)

print(x.shape,y.shape,u.shape,v.shape)
```

https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.quiver.html#matplotlib.axes.Axes.quiver

```python
fig1, ax1 = plt.subplots()
ax1.quiver(x, y, u, v)
```

https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.streamplot.html#matplotlib.axes.Axes.streamplot

```python
fig1, ax1 = plt.subplots()
ax1.streamplot(x, y, u, v)
```

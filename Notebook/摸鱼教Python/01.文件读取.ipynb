数据格式：

1、NDArray:

在numpy中，array是一个函数，用来创建一个矩阵对象,而ndarray是该对象的类型.

2、DataFrame:

DataFrame是一个表格型的数据结构，它含有一组有序的列，每列可以是不同的值类型（数值、字符串、布尔值等）。DataFrame既有行索引也有列索引，它可以被看做由series组成的字典（共用同一个索引）

3、DataArray:

是一个使用标签的多维数组，主要包含以下属性：

1. values：一个保存数组值的numpy.ndarray
2. dims: 每个坐标轴的维度名称 (例如， (‘x’, ‘y’, ‘z’))-
3. coords: 一个包含数组坐标的类似字典的容器，用来标记每个点（例如，数字，日期时间对象或字符串的一维数组）
4. attrs: 存放任意元数据（属性）的字典

## 举栗子

使用函数：pandas.read_csv()

参数：
1. filepath_or_buffer：路径，也可以是URL
2. sep：指定分隔符。如果不指定参数，则会尝试使用逗号分隔，多空格可使用'\s+'
3. header：指定行数用来作为列名，数据开始行数。如果文件中没有列名，则默认为0，否则设置为None。
4. names: 用于结果的列名列表，如果数据文件中没有列标题行，就需要执行header=None。
5. false_values: 设定缺测值
6. skiprows: 设定需要跳过的行号(从前往后)
7. skipfooter : 设定需要跳过的行号(从后往前)
8. skipinitialspace: 是否忽略分隔符后的空格
9. nrows: 从前数需要读取的行号 
10. na_values: 用于替换缺测值
11. parse_dates： 解析日期
12. usecols: 返回一个数据子集，该列表中的值必须可以对应到文件中的位置（数字可以对应到指定的列）

    boolean. True -> 解析索引
    
    list of ints or names. e.g. If [1, 2, 3] -> 解析1,2,3列的值作为独立的日期列；
    
    list of lists. e.g. If [[1, 3]] -> 合并1,3列作为一个日期列使用
    
    dict, e.g. {‘foo’ : [1, 3]} -> 将1,3列合并，并给合并后的列起名为"foo"

# GRIB文件

GRIB文件名通常以.grb、.grib、.grb2等结尾（并不绝对，比如GFS预报以预报时次作为文件后缀）。实际上GRIB文件分为GRIB1和GRIB2两个版本，但是此处暂不做区分。

Xarray支持两种GRIB的文件解析引擎：

	cfgrib：由ECMWF开发，支持Windows、Linux和macOS。对于某些非标准GRIB特征不提供支持。但是因为cfgrib是Windows上的唯一选项，所以Windows用户必须要选cfgrib作为GRIB解析引擎。


	PyNIO：由NCAR开发，来源于NCL，不支持Windows。因为底层与NCL同源，对GRIB的支持非常好，并且对NCEP使用的非标准特征有完善支持，所以在读取大部分GRIB文件时都会有良好表现。但是，PyNIO后续开发已经流产。且与库netCDF4>1.5.3版本冲突，在某些较新Python版本上大量读取时可能出现内存泄漏问题，未来Xarray可能会停止对PyNIO的支持。


```python
dataset = xr.open_dataset('/home/mw/input/moyu4302/example4.grib2',engine='cfgrib')
print(dataset)
```

# Grads二进制文件

Grads的二进制文件名通常以.grb后缀结尾，并带有与之配套的.ctl描述文件。同时支持站点数据和网格数据。

首选建议：使用CDO转为.nc文件

cdo -f nc import_binary infile.ctl outfile.nc

#### åˆ†ç»„èšåˆ
##### ğŸ’¥SQL
  ```sql
  select column_A, sum(column_B)
  from table_name
  group by column_A
  ```
##### ğŸ’¥Pandas
**åŸºæœ¬ç”¨æ³•ï¼š**
å¯¹DataFrameè¿›è¡Œgoupbyè¿ç®—åï¼Œè¿”å›çš„æ˜¯ä¸€ä¸ªgroupbyå¯¹è±¡ï¼Œæˆ‘ä»¬å¯ä»¥é€šè¿‡`.reset_index()`å°†å…¶è½¬ä¸ºDataFrameã€‚

````python
# ä»¥Ownership Typeåˆ—åˆ†ç»„ï¼Œå¯¹Brandåˆ—è¿›è¡Œè®¡æ•°
# .reset_index()å°†groupbyå¯¹è±¡è½¬æˆdataframe
data.groupby(['Ownership Type'])['Brand','Country'].count().reset_index()
````

````python
# ä»¥Countryå’ŒCityåˆ—è¿›è¡Œåˆ†ç»„ï¼Œå¯¹Longitudeè¿›è¡Œæ±‚å¹³å‡
data.groupby(['Country', 'City'])['Longitude'].mean().reset_index()
````

**é«˜é˜¶ç”¨æ³•ï¼š**
æˆ‘ä»¬å¯ä»¥åŒæ—¶å¯¹äºä¸åŒåˆ—é‡‡å–ä¸åŒçš„èšåˆè¿ç®—ï¼Œè­¬å¦‚å¯¹Aåˆ—ä½¿ç”¨`sum()`ï¼Œå¯¹Båˆ—ä½¿ç”¨`mean()`ï¼Œåœ¨SQLä¸­å…¶å®å¾ˆå¥½å®ç°çš„åŠŸèƒ½ï¼Œåœ¨Pandasæˆ‘ä»¬éœ€è¦å€ŸåŠ©`.agg()`æ¥å®ç° ã€‚

````python
# å¯¹ä¸åŒåˆ—è¿›è¡Œä¸åŒå¯¹è¿ç®—
# å¯¹Longitudeè¿›è¡ŒMAXæ“ä½œï¼Œå¯¹Cityåˆ—è¿›è¡ŒCount
data.groupby(['Ownership Type']).agg({'Longitude':'max', 'City':'count'}).reset_index()
````

````python
# å¯¹ç»Ÿä¸€åˆ—è¿›è¡Œä¸åŒå¯¹æ“ä½œ
data.groupby(['Ownership Type'])['Longitude'].agg(['max',   'count']).reset_index()
````

#### è¿æ¥
##### ğŸ’¥SQL
```sql

```
##### ğŸ’¥Pandas
åœ¨Pandasä¸­æˆ‘ä»¬å¯ä»¥ä½¿ç”¨`pandas.merge()`æ¥å®Œæˆè¿æ¥å¯¹æ“ä½œã€‚

 ```python
  pandas.merge(left, right, how='inner', on=None, left_on=None, 
            right_on=None, left_index=False, right_index=False, sort=True)
  ```
å„å‚æ•°è§£é‡Šå¦‚ä¸‹ï¼š
 * leftï¼šä¸€ä¸ªDataFrameå¯¹è±¡ï¼›
 * rightï¼šå¦ä¸€ä¸ªDataFrameå¯¹è±¡ï¼›
 * howï¼šè¿æ¥æ–¹å¼ï¼Œé»˜è®¤ä¸ºinnerï¼ˆå†…è¿æ¥ï¼‰ï¼›
 * onï¼šè¿æ¥é”®ï¼Œå¿…é¡»åœ¨leftå’Œrightä¸¤ä¸ªDataFrameä¸­å­˜åœ¨ï¼Œå¦åˆ™ä½¿ç”¨left_onå’Œright_on;
 * left_onï¼šleftä¸­çš„è¿æ¥é”®ï¼›
 * right_onï¼šrightä¸­çš„è¿æ¥é”®ï¼›
 * left_index/right_indexï¼šé»˜è®¤ä¸ºFalseï¼Œå¦‚æœä¸ºTrueåˆ™ä½¿ç”¨ç´¢å¼•ä½œä¸ºè¿æ¥çš„é”®ã€‚

````python
# ç”Ÿæˆä¸¤ä¸ªDataFrame
left = pd.DataFrame({
         'id':[1,2,3,4,5],
         'Name': ['Alex', 'Amy', 'Allen', 'Alice', 'Ayoung'],
         'subject_id':['sub1','sub2','sub4','sub6','sub5']})
right = pd.DataFrame(
         {'id':[1,2,3,4,5],
         'Name': ['Billy', 'Brian', 'Bran', 'Bryce', 'Betty'],
         'subject_id':['sub2','sub4','sub3','sub6','sub5']})

df = pd.merge(left=left, right=right, on='subject_id', how='left')
df
````

#### æ­£åˆ™è¡¨è¾¾å¼
##### ğŸ’¥SQL
ç›®å‰å¤§éƒ¨åˆ†æ•°æ®åº“éƒ½æ˜¯æ”¯æŒæ­£åˆ™è¡¨è¾¾å¼çš„ï¼Œå› ä¸ºä¸åŒæ•°æ®åº“å¯¹åº”çš„sqlè¯­æ³•ä¸ä¸€è‡´ï¼Œè¿™è¾¹å°±ä¸ä¸¾ä¾‹äº†ã€‚
##### ğŸ’¥Pandas

````python
# ç”Ÿæˆä¸€ä¸ªDataFrame
df = pd.DataFrame({
         'Name': ['Alex', 'Amy', 'Allen', 'Alice', 'Ayoung'],
         'Location':['æ¹–åŒ—çœæ­¦æ±‰å¸‚','å¹¿ä¸œçœæ·±åœ³å¸‚','å¹¿ä¸œçœå¹¿å·å¸‚','æ¹–å—çœé•¿æ²™å¸‚','æ¹–åŒ—çœé„‚å·å¸‚']})

df
````

````python
# è¿”å›ä¸€åˆ—çš„æ—¶å€™expandä¸ºTrueè¿”å›çš„æ˜¯Dataframe
df['Location'].str.extract('(.*?)çœ', expand=True)
````

````python
# è¿”å›ä¸€åˆ—çš„æ—¶å€™expandä¸ºTrueè¿”å›çš„æ˜¯Dataframe
df['Location'].str.extract('(.*?)çœ(.*?)å¸‚', 
                            expand=True).rename(columns = {0: 'Province', 1: 'City'})
````

````python
# ä¸åŸDataFrameæ‹¼æ¥ï¼Œaxis=1è¡¨ç¤ºä¸ºæ¨ªå‘æ‹¼æ¥
pd.concat([df, df['Location'].str.extract('(.*?)çœ(.*?)å¸‚', expand=True
                                ).rename(columns = {0: 'Province', 1: 'City'})], axis=1)
````

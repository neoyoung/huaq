## 2.9 缺失值nan的提取与处理

```python
# # 选取数据框的部分数据
# 选取数据框中有nan的列
df_part = df[(df['sex'].isna()) | (df['amount'].isna())]
df_part
```

```python
# # 选取数据框的部分数据
# 选取数据框中没有nan的列
df_part = df[~((df['sex'].isna()) | (df['amount'].isna()))]
df_part

```

```python
# 上面是知道了这两列有nan，所以这么操作，那如果数据很大呢，该怎么操作呢？
df.isnull().any()
```

```python
# 当我们有上面的基础信息后，就可以把为空的列找出来
# 比如使用index和values的方式，结合zip函数
nan_cols = []
for f1, f2 in zip(df.isnull().any().index, df.isnull().any().values):
    if f2 == True:
        nan_cols.append(f1)
print('nan cols: ', nan_cols)

# 也可以这么写，使用reset_index(drop = False), 即不删除索引，那么索引就会当成列
nan_df = df.isnull().any().reset_index(drop = False)
nan_df.columns = ['columns_name', 'is_nan'] # 对新的两列重新命名（不命名也行，只是为了看的更清楚）
nan_cols2 = nan_df[nan_df['is_nan'] == True]['columns_name'].values # 可以对这里拆分开来写
nan_cols2
```

```python
# 为空的行怎么确定呢, 可以对数据框转置(即，df.T), 为空的行的索引会显示True
df.T.isnull().any()
```

```python
# 那想把缺失的行, 数据框展示出来呢
df[df.T.isnull().any() == True]
```

```python
# 那如果要统计一列的缺失率呢？
# 可以使用count方法, 因为count计算的是非空的个数，因此如果要计算sex列的缺失率
print('sex nan ratio ', df['sex'].count() / len(df))

# 当然我们也可以写个函数遍历所有的列，看看缺失率
def get_nan_ratio(col, df):
    return df[col].count() / len(df)

for f in df.columns:
    ratio = get_nan_ratio(f, df)
    print(f'{f} nan ratio: ', ratio)
```

```python
# # 选取数据框的部分数据
# 选取数据框中指定列的值，比如数据框中年龄指定为23，12，26，9
# 我们可以这么写, 但是其实写的很长，如果指定的范围更多呢
df_part = df[(df['age'] == 23) | (df['age'] == 12) | (df['age'] == 26) | (df['age'] == 9)]
df_part
```

## 2.11 重塑索引reset_index()

```python
# 可以看到df_part的索引不是从0开始了，我们可以用reset_index(drop = True), 重新构建索引（就是把原来的索引删除，因为drop = True的作用，然后生成一个默认从0开始的索引）
# 这样做的原因是因为，对于选取数据框中的数据，
# 比如iloc，是按照索引选取的，当你没意识到索引被改变了，还按照最初想法选取索引数据，那么就会取错数据
df_part = df_part.reset_index(drop = True)
df_part
```

## 2.13 map方法，新增一列

## 2.14 astype方法，修改series类型（附带replace方法）

```python
# 上面的sex_map 中的数字 其实是字符串
print(df['sex_map'][0], type(df['sex_map'][0]))

# 如何把他转化成整数呢
# 可以使用astype方法
'''
.astype(float)转为浮点型
.astype(string) 转为字符串 （一般拼接字段的时候需要）
''' 
# 但是直接使用下面这个代码会报错，因为 sex_map 列里面包含缺失值
# df['sex_map'] = df['sex_map'].astype(int)

# 那么就需要先把缺失值填充了，使用fillna方法，比如对这一列填充 ‘-1’
df['sex_map'] = df['sex_map'].fillna('-1')

# 然后就可以使用 astype 方法了
df['sex_map'] = df['sex_map'].astype(int)
# 可以看到， sex_map 里的值已经被转为int了
print(df['sex_map'][0], type(df['sex_map'][0]))

# 这个时候可以使用 replace 方法再给替换回去
'''
replace(要替换的变量，替换后的变量)
'''
df['sex_map'] = df['sex_map'].replace(-1, np.nan)
df
```

## 2.15 groupby方法，聚合操作

```python
# 高阶用法
# 有时候，我们想统计某个维度1下维度2的总和
# 比如，想看不同性别下的收入总和
temp = df.groupby(['sex'])['amount'].sum()
print(type(temp))
temp
```

```python
# groupby聚合之后，temp变成了series， 
# 我们将series变成数据框, 同时我们想把age这个索引保留：
temp = temp.reset_index(drop = False)
temp
```

```python
# 聚合成上面数据框之后，我们可以看到amount列真实的含义其实是 amount的求和，因此为了避免混淆，我们可以重新命名
temp.columns = ['sex', 'amount_sum']
temp
```

## 2.16 merge方法，数据拼接

```python
# 我们想把temp拼接到原来的df上呢（因为他们共同有age这个主键字段），那就可以使用merge方法
# 以age为主键，how = 'letf' 表示以左边的数据框作为主表，拼接上temp
df = df.merge(temp, on = 'sex', how = 'left')
df
```

## 2.17 两个维度聚合、拼接

```python
# 有时候，我们想统计 多个维度下 另一个维度 的 总和
# 比如，想看不同性别下、不同年龄的收入总和，这个时候二阶统计就可以，
# 由于上面演示过整个流程，这里直接一个cell写了
temp = df.groupby(['sex', 'age'])['amount'].sum().reset_index(drop = False)
temp.columns = ['sex', 'age', 'amount_sum2']
df = df.merge(temp,  on = ['sex', 'age'], how = 'left')
df
```

## 2.18 agg方法，统计更多信息

```python
# 当我们想统计 一个维度下，另一个维度的总和，均值，最大值，最小值，中位数；
# 我们用上面的方法一个个去写其实是很麻烦的，这个使用可以使用agg方法
# 比如想统计，不同性别下，年收入的总和，均值，最大值，最小值，中位数；年龄的总和，均值，最大值，最小值，中位数；
# 先创建一个下面的字典
sex_amount_agg_stats = {'amount': ['mean', 'max', 'min', 'median'], # 因为前面计算过sum了，这里就不写了，否则重复计算
                        'age': ['sum', 'mean', 'max', 'min', 'median'],} 
temp = df.groupby(['sex']).agg(sex_amount_agg_stats)
print(type(temp))
temp
```

```python
# 我们先看这个长相复杂的数据框temp 的列是哪些
print('列： ', temp.columns)

# 我们先看这个长相复杂的数据框temp 的索引是什么
print('索引： ', temp.index)
```

```python
# 遍历打印temp的列呢，发现是元组
for f in temp.columns:
    print(f)
```

```python
# 因此我们可以直接用这个列，来构建新的列名
temp.columns = [f[0] + '_' + f[1] for f in temp.columns]
temp
```

```python
# 这个时候就可以重塑索引了, 保留sex索引
temp = temp.reset_index(drop = False)
temp
```

```python
# 那么就可以按照sex，拼接到sex上了
df = df.merge(temp, on = 'sex', how = 'left')
df
```

## 2.19 transform方法，新增一列的最爱

```python
# 如果我不想为了新增一个列，做了一堆的修改列名、reset_index索引、merge拼接，
# 那么我们可以使用transform方法
# 比如我想统计 sex 的 不同性别的个数, 并新增一列 sex_count
# 这里的name_id 也可以用其他的字段，因为是统计个数，所以这个字段最好不要是空
df['sex_count'] = df.groupby('sex')['name_id'].transform('count')
df
```

其实上面的聚合、拼接过程，就是产生新的列（我们也把列叫做特征）的过程，我们把这个过程叫做特征工程

## 2.20 concat方法，数据拼接

```python
# 可以看到merge方法，帮助两个数据框，通过主键，实现左右连接（how = 'right' 右连接）
# 当面我们想上下拼接呢，那么可以使用 concat 方法；
# 由于concat方法不需要主键
# 所以concat 左右/上下拼接，从数据质量角度来说，是要求左右/上下拼接之后的，每条数据是有意义的，不然实现了拼接，但是没有实际意义

# 我们先分出两份数据框
df_man   = df[df['sex'] == '男'].reset_index(drop = True)
df_woman = df[df['sex'] == '女'].reset_index(drop = True)

df_man
```

```python
df_woman
```

```python
# 我们就可以把上面的两份数据框上下拼接, 注意axis = 0，是上下拼接
df = pd.concat([df_man, df_woman], axis = 0).reset_index(drop = True)
df
```

```python
# concat 中 axis = 1，是左右拼接，当行数不同的时候，左右拼接，就会出现问题（虽然没报错，但是数据已经没有意义了）
df_axis1 = pd.concat([df_man, df_woman], axis = 1).reset_index(drop = True)
df_axis1
```

```python
# 我们将数据进行下面切分，然后左右拼接，数据是有意义的
df_raw1 = df[['name', 'age', 'sex', 'amount', 'sex_map', 'name_id']]
df_raw2 = df[['amount_sum','amount_sum2', 'sex_mean', 'sex_max', 'sex_min', 'sex_median']]

df_raw1
```

```python
df_raw2
```

```python
# 我们就可以把上面的两份数据框上下拼接, 注意axis = 1，是左右拼接
df = pd.concat([df_raw1, df_raw2], axis = 1)
df
```

## 2.21 drop方法，数据框指定列的删除

```python
# 在上面左右数据框切分的时候，我们通过指定列名，也可以对数据框切分
# 那么如果有一列或者几列真的是我们不需要的呢，那就可以使用drop方法了
# 比如删除 amount_sum2 、 sex_median 这两列数据, 
# 注意axis = 1 表示列，inplace = True才可以将原来的df 替换为删除后的 df
df.drop(['amount_sum2', 'sex_median'], axis = 1, inplace = True)
df
```

```python
# 上面说到axis = 1,是删除列；那么删除行呢
# 比如想删除第三行数据

df.drop(index = [2], axis = 0, inplace = True)
# 当我们想批量删除有某些特点的行，那就可以先想想怎么操作把相应的索引找到，就很好删除了

df
```

## 2.22 drop_duplicates方法，删除重复数据

```python
# drop_duplicates
# 默认检验所有的行是否重复， 注意inplace = True，才可以将df 替换为 修改后的df
df.drop_duplicates(inplace = True)
df
```

```python
# 如果想删除指定列的行有重复呢
# 比如我们之前说到的外国学生和国内学生，我们这两个数据框拼接起来之后，只想要所有学生名单数据框
# 先构建这样重复的数据框（列的顺序不同没事，因为concat可以自己调整列的顺序，只要保证列名相同就行）
df = pd.concat([df, df_usa], axis = 0).reset_index(drop = True)
df
```

```python
# 由于只想要全部学生名单，因此我们可以只按照 name 字段来过滤重复的
# keep = 'first' 表示重复的数据，取第一个；
'''
keep： 可选参数有三个：first、 last、 False， 默认值 first。其中:
first 表示： 保留第一次出现的重复行，删除后面的重复行
last  表示： 删除重复项，保留最后一次出现
False 表示： 删除所有重复项
'''
df.drop_duplicates(subset=['name'], keep='first', inplace=True)
df
```
